```python
# Importing Python libraries
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib notebook
import math

# naive_bayes lilbraries
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
dataset = pd.read_csv('buy_land_data.csv')
```

```python
dataset
```

```python
# drop fields
dataset = dataset.drop('NO', axis=1)
dataset
```

```python
# convert dataset to numeric since data given is numeric data
# Because the data used is categorical data and the model 
# only understands numerical data. We need to convert the data 
# from categories to numeric using cat.codes or you can also use labelencoder.

# convert datatype object to category
dataset_convert = dataset.astype('category')
```

```python
dataset_convert
```

```python
# convert categories to numeric
dataset_convert["INCOME"]        = dataset_convert["INCOME"].cat.codes
dataset_convert["AGE"]           = dataset_convert["AGE"].cat.codes
dataset_convert["STUDENT"]       = dataset_convert["STUDENT"].cat.codes
dataset_convert["CREDIT_RATING"] = dataset_convert["CREDIT_RATING"].cat.codes
dataset_convert["BUYS_LAND"]     = dataset_convert["BUYS_LAND"].cat.codes
```

```python
# Dataset converted to numeric data
dataset_convert
```

```python
# split the labels and the features data
X = dataset_convert.iloc[:,:4].values
y = dataset_convert.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
# Splitting the dataset into training set and test set 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=45) 
```

```python
print(X_train.shape, y_train.shape)
```

```python
# Train the model with predefined parameters
# create gausian classifier
model = GaussianNB()
```

```python
# train models with training data
model.fit(X_train,y_train)
```

```python
# response prediction for test data
y_pred = model.predict(X_test)
```

```python
# lets make the prediction 
# Predict : age<=30 = 1 , income = medium = 2, student = yes = 1 , credit_rating = fair = 1
predicted = model.predict([[1,2,1,1]]) 
```

```python
print("Predicted Value is : " , predicted)
```

```python
# accuracy score
print("Accuracy Score is  :" ,  metrics.accuracy_score(y_test, y_pred))
```

```python
# create a confussion matrixfrom sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix is : ", confusion_matrix)
```

From the code representation , since the Output of the predicted value is : 1 , which represents a YES occurance, then we would say that the prediction of buys_land when given conditions : age<=30, income=medium, student=yes, credit-rating=fair ,is YES and the accuracy score is 100%
